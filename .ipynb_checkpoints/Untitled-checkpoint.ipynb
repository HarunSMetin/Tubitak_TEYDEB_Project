```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


#import seaborn as sns
import pandas as pd

from scipy.optimize import minimize_scalar;


#from lmfit.models import LinearModel, LorentzianModel
```

```python

# Define the Lorentzian function
def lorentzian(x, x0, gamma, A):
    return A * (gamma**2 / ((x - x0)**2 + gamma**2))
```

```python
df = (pd.read_excel("deneme.xlsx"))
column = len(df.iloc[0,:])
row = len(df.iloc[:,0])
df
```

```python

h=475
#df=df[df.NM<(h+50)]
#df=df[df.NM>(h-50)]

#initial_guess = [0.1, 0.1, 2.0]
A = 0.09463919098664426
x0 = 274.3146806018545
gamma = 585.8269681890237 

initial_guess = [ A,x0, gamma]
x = df.NM.to_numpy()
y_data =df.iloc[:,1].to_numpy() 

fit_params, _ = curve_fit(lorentzian, x, y_data, p0=initial_guess,maxfev=40000000)

# Generate the fitted curve
y_fit = lorentzian(x, *fit_params)

print("AR MAX ",df.NM.iloc[y_fit.argmax()])

# Plot the data and the fitted curve
plt.figure()
plt.plot(x, y_data, 'g.', label='Data')
plt.plot(x, y_fit, 'b', label='AR')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def lorenzian_derivative(x, A, x0, gamma):
    numerator = 2 * A * gamma**2 * (x - x0)
    denominator = ((x - x0)**2 + gamma**2)**2
    return numerator / denominator

# Fonksiyonun tepe noktası olduğu düşünülen x değerleri
x_values = df.NM.iloc[y_fit.argmax()-40:y_fit.argmax()+40] # Bu listeye uygun x değerlerini ekleyin
# Türevi sıfır olan noktaları bulun
for x in x_values:
    derivative_at_x = lorenzian_derivative(x, A, x0, gamma)
    if abs(derivative_at_x) < 0.0001: # Küçük bir eşik değeri belirleyebilirsiniz
        # Fonksiyonun türevi sıfır olduğu noktayı bulduk
        # Şimdi bu noktada fonksiyonun değerini hesaplayarak global maksimumu bulalım
        maximum_value = (A * gamma**2) / (gamma**2)
        print(f"Global maksimum değer: {maximum_value} at x={x}")
```
