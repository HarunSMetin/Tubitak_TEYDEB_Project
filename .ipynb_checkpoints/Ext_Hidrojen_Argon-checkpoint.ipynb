```python
#import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Chebyshev as T
#from scipy.optimize import minimize_scalar

def find_nearest_index(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

def findMax(array): 
    maxIndex = 0
    for i in range(0,len(array)): 
        max = float(array[i].real) 
        if max > 2:
            continue
        if array[i]>max:
            max = array[i]
            maxIndex = i
    return (maxIndex-1,array[maxIndex-1])
```

```python
def MakeGrafs(df,degree,show : False ,ofset:50):
    x = df.NM.to_numpy()
    PolyDf = df.copy()
    listOfPeaks = []
    for c in df.columns:
    # Örnek veri oluşturma
        y =df[c].to_numpy()
        #Grafik figürünü oluşturun
        
        # Polinomial Regresyon modelini uygulama

        poly = np.poly1d(np.polyfit(x, y, degree))
        """
        # Polinomun tepe noktalarını bulma
        derivative= np.polyder(poly)
        max_x_array = np.roots(derivative) 
        a,max_y=findMax(poly(max_x_array) )
        max_x=max_x_array[a] 
               
        """
        # ARGON Fonksiyonun negatifini alarak maksimum değeri bulma
        neg_poly = lambda x: -poly(x)
        result = minimize_scalar(neg_poly)

        # ARGON Mutlak maksimum değeri ve konumunu bulma
        max_x = result.x
        max_y = poly(max_x)
        
        peak = (max_x.real,max_y.real)
        listOfPeaks.append(max_x.real) 
        PolyDf[c] = poly(x)
        # Grafik oluşturma   
        if show==True:
            plt.figure(figsize=(15,15))

            plt.plot(x,y, label=c ,color='orange',) 

            plt.plot(x, poly(x), color='blue', label=c)

            # Mutlak maksimumu işaretleme
            plt.plot (max_x,max_y, marker='o', markersize=8, color='green',label=('Mutlak Maksimum '+c))
            plt.axvline(x=max_x, color='green', linestyle='--')

            plt.xlabel('nm')
            plt.ylabel('abs')
            plt.legend()
            plt.show()
    return listOfPeaks,PolyDf
```
