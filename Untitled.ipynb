```python

# Define the Lorentzian function
def lorentzian(x, x0, gamma, A):
    return A * (gamma**2 / ((x - x0)**2 + gamma**2))
```

```python
df = (pd.read_excel("deneme.xlsx"))
column = len(df.iloc[0,:])
row = len(df.iloc[:,0])
df
```

```python
def lorenzian_derivative(x, A, x0, gamma):
    numerator = 2 * A * gamma**2 * (x - x0)
    denominator = ((x - x0)**2 + gamma**2)**2
    return numerator / denominator

# Fonksiyonun tepe noktasÄ± olduÄŸu dÃ¼ÅŸÃ¼nÃ¼len x deÄŸerleri
x_values = df.NM.iloc[y_fit.argmax()-40:y_fit.argmax()+40] # Bu listeye uygun x deÄŸerlerini ekleyin
# TÃ¼revi sÄ±fÄ±r olan noktalarÄ± bulun
for x in x_values:
    derivative_at_x = lorenzian_derivative(x, A, x0, gamma)
    if abs(derivative_at_x) < 0.0001: # KÃ¼Ã§Ã¼k bir eÅŸik deÄŸeri belirleyebilirsiniz
        # Fonksiyonun tÃ¼revi sÄ±fÄ±r olduÄŸu noktayÄ± bulduk
        # Åžimdi bu noktada fonksiyonun deÄŸerini hesaplayarak global maksimumu bulalÄ±m
        maximum_value = (A * gamma**2) / (gamma**2)
        print(f"Global maksimum deÄŸer: {maximum_value} at x={x}")
```
